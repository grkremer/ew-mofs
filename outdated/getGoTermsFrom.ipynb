```python
import requests
from xml.etree import ElementTree

def get_accession_synonyms(accession):
    # URL base da API do NCBI
    base_url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/"
    
    # Usar EFetch para obter os metadados associados ao accession
    efetch_url = f"{base_url}efetch.fcgi"
    efetch_params = {
        "db": "nuccore",  # Banco de dados de sequências
        "id": accession,  # O accession fornecido
        "rettype": "docsum",  # Retornar um resumo dos metadados
        "retmode": "xml"
    }
    efetch_response = requests.get(efetch_url, params=efetch_params)
    
    # Parse do XML retornado
    tree = ElementTree.fromstring(efetch_response.content)
    
    # Coletar sinônimos do accession
    synonyms = []
    for item in tree.findall(".//Item[@Name='Extra']"):
        data = item.text
        if data:
            for entry in data.split(";"):
                if entry.startswith("Accession"):
                    synonym_list = entry.split("=")[-1].split(",")
                    synonyms.extend(synonym_list)
    
    if synonyms:
        return list(set(synonyms))  # Retorna sinônimos únicos
    else:
        return f"Sinônimos não encontrados para o Accession: {accession}"

# Exemplo de uso
accession = "XR_108906"
result = get_accession_synonyms(accession)
print(f"Sinônimos para {accession}: {result}")
```

```python
import requests

def get_go_terms_from_gene_symbol(gene_symbol):
    # URL base da API MyGene.info
    base_url = "https://mygene.info/v3/query"
    
    # Parâmetros da consulta
    params = {
        "q": gene_symbol,        # Gene Symbol
        "fields": "go",          # Campo para termos GO
        "species": "human",      # Espécie (altere conforme necessário)
        "size": 1                # Limitar a 1 resultado mais relevante
    }
    
    # Requisição à API
    response = requests.get(base_url, params=params)
    
    if response.status_code != 200:
        return f"Erro ao consultar API: {response.status_code}"
    
    data = response.json()
    
    # Verificar se existem dados de GO
    if "hits" not in data or len(data["hits"]) == 0:
        return f"Nenhuma anotação GO encontrada para o gene {gene_symbol}"
    
    # Obter termos GO (processos biológicos, funções moleculares, componentes celulares)
    go_data = data["hits"][0].get("go", {})
    go_terms = {
        "biological_process": go_data.get("BP", []),
        "molecular_function": go_data.get("MF", []),
        "cellular_component": go_data.get("CC", [])
    }
    
    # Formatando os resultados
    result = {}
    for category, terms in go_terms.items():
        result[category] = [f"{term['id']}: {term['term']}" for term in terms]
    
    return result

# Exemplo de uso
gene_symbol = "RFC2"
go_terms = get_go_terms_from_gene_symbol(gene_symbol)
print(f"Termos GO para o gene {gene_symbol}:\n")
for category, terms in go_terms.items():
    print(f"{category.capitalize()}:")
    for term in terms:
        print(f"  - {term}")
'''
0000902 // cell morphogenesis // inferred from electronic annotation /// 
0006200 // ATP catabolic process // inferred from direct assay /// 
0006950 // response to stress // inferred from electronic annotation /// 
0006986 // response to unfolded protein // traceable author statement /// 
0034605 // cellular response to heat // inferred from direct assay /// 
0042026 // protein refolding // inferred from direct assay /// 
0070370 // cellular heat acclimation // inferred from mutant phenotype
'''
```
